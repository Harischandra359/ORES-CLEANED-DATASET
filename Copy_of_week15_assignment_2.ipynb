### 3. Aggregate your data frame

Do some kind of aggregation on your data frame.  Something that makes sense and has some groups to it.  Don't just sum up one column for the entire data frame.  Be more creative than that.

```python
#gives the average for value column for every city
city_avg = df.groupby('CITY')['VALUE'].mean()
city_avg
```

```python
# Most common pollutant
most_frequent_pollutant = df['POLLUTANT'].mode().values[0]
print("Most Pollutant in all cities:", most_frequent_pollutant)
```

```python
# Most common pollutant in every city
most_frequent_pollutant_by_city = df.groupby('CITY')['POLLUTANT'].apply(lambda x: x.mode().values[0])
print("Most commom pollutant in every city")
most_frequent_pollutant_by_city
```

```python
# Find the index of the row with the maximum value in the 'VALUE' column
max_value_index = df['VALUE'].idxmax()

# Data with the maximum value
data_of_max_value = df.loc[max_value_index]

print("Most pollutant location\n",data_of_max_value)
```

### 4. Write to Excel

Use Pandas functions to write your summarized data out to a local Excel file.

```python
import pandas as pd

# Specify the Excel file name
excel_filename = 'air_quality.xlsx'

# Write the DataFrame to Excel file
df.to_excel(excel_filename, index=False)

#verification
data_from_excel = pd.read_excel(excel_filename)
data_from_excel
```
